```python
import pandas as pd
import pandera as pa
```

```python
df = pd.read_csv('https://storage.googleapis.com/datalake-soulcode-ed7/arquivos%20soulcode/despesa_alesp.csv')
dfbackup = df.copy()
pd.set_option('display.max_rows', 50)
```

```python
# renomeando as colunas para tudo minúsculo
df.rename(columns={'Ano':'ano','Matricula':'matricula','Mes':'mes','Valor':'valor','CNPJ':'cnpj','Tipo':'tipo', 'Fornecedor':'fornecedor'},inplace=True)
```

```python
# exemplo de tratamento de dados com pedágio
df.replace(['CUPONS DE PEDÁGIO','PEDAGIO','PEDAGIO DE ELISIARIO','PEDÁGIOS','PEGAGIO'],'PEDAGIO',inplace=True)
```

```python
df.dtypes
```

```python
# Criando um schema de validação de dados
schema = pa.DataFrameSchema(
  columns={
    'ano':pa.Column(pa.Int),
    'matricula':pa.Column(pa.Int),
    'mes':pa.Column(pa.Int),
    'valor':pa.Column(pa.Float),
    'cnpj':pa.Column(pa.Float,nullable=True),
    'tipo':pa.Column(pa.String),
    'fornecedor':pa.Column(pa.String),
    }
)
```

```python
# validando o DF
schema.validate(df)
```

```python
# pegando os dados da última legislatura
dfleg = df[(df.ano > 2018) & (df.ano < 2023)]
dfleg
```
